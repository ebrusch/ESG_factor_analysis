```python
#os.chdir("/Users/emiliebrusch/Documents/Speciale/Kode")
ESG_data = pd.read_csv('ESG_P_MV_18Y.csv', sep = ';', decimal = ",")

ff_factor3_data = ff_data.FamaFrenchReader('F-F_Research_Data_Factors', start = pd.Timestamp('2004-06'),
                                                       end = pd.Timestamp('2021-11')).read()[0]

ff_factor4_data = ff_data.FamaFrenchReader('F-F_Momentum_Factor', start = pd.Timestamp('2004-06'),
                                                       end = pd.Timestamp('2021-11')).read()[0]

ff_factor5_data = ff_data.FamaFrenchReader('F-F_Research_Data_5_Factors_2x3', start = pd.Timestamp('2004-06'),
                                                              end = pd.Timestamp('2021-11')).read()[0]


ff_factor3_data = ff_factor3_data.apply(lambda x: x/100, axis=1)
ff_factor4_data = ff_factor4_data.apply(lambda x: x/100, axis=1)
ff_factor5_data = ff_factor5_data.apply(lambda x: x/100, axis=1)
```

```python
col = ESG_data.columns[:-1]
ESG_data = ESG_data.drop('01.04.2004',1)
ESG_data.columns = col
ESG_data
```

# Fordel porteføljer

```python
# extract the data from june every year
june_data = ESG_data[ESG_data.columns[3::12]]
june_data['Name'] = ESG_data['Name']
june_data = june_data[['Name',*ESG_data.columns[3::12]]]
```

```python
# get esg scores from each company
esg_idx = june_data['Name'].str.split('-').str[-1].str.strip(' ') == 'ESG Score'
esg_june_data = june_data[esg_idx]
esg_june_data

# get MV scores from each company
esg_mv_idx = june_data['Name'].str.split('-').str[-1].str.strip(' ') == 'MARKET VALUE'
mv_june_data = june_data[esg_mv_idx]
assert(len(esg_june_data) == len(mv_june_data))
```

```python
# get list of companies in each portfolio each year
# at each key in all of these portfolios, there should exist a new dictionary.
# keys are decile numbers
companies = dict()
esg = dict()
meanesg = dict()

# loop through all year in the dataset
for year in esg_june_data.columns[1:]:
    # get all companies for which the ESG Scores are not NaN
    company_names = esg_june_data['Name'][esg_june_data[year].notnull()].str[:-12]
    
    # divide into 10 partitions based on ESG Scores
    deciles = pd.qcut(esg_june_data[year][esg_june_data[year].notnull()], 10, labels=False)
    
    # Loop through every decile and assign each decile to the companies dictionary
    for dec in range(10):
        dec_idx = deciles == dec
        dec_names = company_names[dec_idx]
        
        # make sure the dictionarys exist at each decile key - and use 1 indexing for good measure.
        if dec+1 not in companies.keys():
            companies[dec+1] = dict()
            esg[dec+1] = np.array([])
            meanesg[dec+1] = dict()
            
        # fill decile keys with data
        esg[dec+1] = np.append(esg[dec+1], esg_june_data[year][esg_june_data[year].notnull()][dec_idx].values)
        meanesg[dec+1][year] = np.mean(esg_june_data[year][esg_june_data[year].notnull()][dec_idx].values)

        companies[dec+1][year] = dec_names
```

# Oversigt over porteføljer
Ændre tal og årstal for at se antal virksomheder osv. i hvert portefølge for hvert år

```python
# ESG & MV
# 
stats_dict_esg = dict()
stats_dict_mv = dict()
for k in companies:
    stats_dict_mv[k] = dict()
    stats_dict_esg[k] = dict()
        
for i in companies:
    for year in companies[i]:
        
        names = companies[i][year] + " - MARKET VALUE"
        stats_dict_mv[i][year] = ESG_data.loc[ESG_data["Name"].isin(list(names))][year].sum()
        
        names = companies[i][year] + " - ESG Score"
        stats_dict_esg[i][year] = ESG_data.loc[ESG_data["Name"].isin(list(names))][year].mean()
```

```python
# ændrer tal for at ændre portefølje. Ændre årstal
portfolio = 10
year = '2020'
print('Antal virksomheder:\n', len(companies[portfolio]['01.06.' + year]))
print('\nVirksomheder:\n', companies[portfolio]['01.06.' + year].values)
print('\nMiddel ESG for portefølje i givne år:',meanesg[portfolio]['01.06.' + year])
print('\nMiddel ESG for portefølje på tværs af år:', np.mean(esg[portfolio]))
print('\nMiddel MV for portefølje på tværs af år:',np.mean(list(stats_dict_mv[portfolio].values())))
print('\nSum MV for portefølje på tværs af år:', np.sum(list(stats_dict_mv[portfolio].values())))
```

# Regn afkast for alle virksomheder

```python
# afkast
afkast = ESG_data.iloc[0::6]
afkast.index = afkast['Name']
afkast = afkast.drop(columns='Name')
afkast = afkast.T
afkast = afkast.astype('float64')

# market value
MV = ESG_data.iloc[1::6]
MV.index = MV['Name']
MV = MV.drop(columns='Name')
MV = MV.T
MV = MV.astype('float64')

#use columns names from afkast
MV.columns = afkast.columns
```

```python
afkast_diff = afkast.pct_change(1)
```

# Construct portfolios

```python
portfolio = pd.DataFrame()
VALUE_WEIGHTED = True

for dec in companies.keys():
    dec_temp = pd.DataFrame()
    for year in companies[dec].keys():
        portfolio_dec_year = companies[dec][year]
        start_idx = np.where(afkast_diff.index == year)[0][0]
        
        # afkast
        period_afkast = afkast_diff[start_idx:start_idx+12][portfolio_dec_year]
        
        if VALUE_WEIGHTED:
            #weights
            period_MV = MV[start_idx:start_idx+12][portfolio_dec_year]
            period_MV = period_MV.apply(lambda row: row[list(period_MV.columns)]/row.sum(), axis=1)
        
            # weighted returns based on MV
            period_afkast = period_afkast.multiply(period_MV, axis=1)
            
            # take the mean on the weighted returns
            data = period_afkast.sum(axis=1)
        else:
            data = period_afkast.mean(axis=1)
        

        temp = pd.DataFrame({dec: data}, index = afkast_diff.index[start_idx:start_idx+12])
        dec_temp = pd.concat((dec_temp, temp), axis = 0)
        
    portfolio[dec] = dec_temp


```

```python
def linreg(Y,X):
    X = sm.add_constant(X, prepend=True)
    mod = sm.OLS(Y,X)
    res = mod.fit()
    return res
```

```python
### CAPM
portfolio_navn = "GMB" # 1..10 + "GMB"

CAPM_X = np.array(list(ff_factor3_data['Mkt-RF']))
CAPM_Y = np.array(list(portfolio_merafkast[portfolio_navn]))

print(CAPM_Y.shape)
res = linreg(CAPM_Y, CAPM_X)

print(res.summary())
```
